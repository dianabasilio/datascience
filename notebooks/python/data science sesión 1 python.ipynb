```python
#Voy a eliminar aquella(s) variable(s)/(columnas) que contienen muchos NULL y que no son de utilidad para mi estudio, en este caso "company"
df_clean = df.drop("company", axis=1).copy()
df_clean
```

```python
#Si deseo reemplazar/sustituir algÃºn NULL, por ejemplo reemplazar los NULL por ceros en la variable "children"
#df_clean["children"] = df_clean["children"].fillna(0)
df_clean["children"].fillna(0, inplace=True)
df_clean
```

```python
#Hago lo mismo que en el paso anterior para la variable "days_in_waiting_list"
# df_clean["days_in_waiting_list"] = df_clean["days_in_waiting_list"].fillna(0)
df_clean["days_in_waiting_list"].fillna(0, inplace=True)
df_clean
```

```python
#Elimino todas las filas que contengan algÃºn NULL
# df_clean = df_clean.dropna()
df_clean.dropna(inplace=True)
df_clean
```

```python
#Contabilizamos cuantos NULL tiene nuestra tabla
df_clean.isnull().sum()
```

```python
#NUESTRA TABLA HA QUEDADO LISTA SIN NULL!

#Ahora vamos a normalizar los datos,
#para ajustar los valores medios en diferentes escalas respecto a una escala comÃºn
#ya que hay variabes que contienen datos muy grandes respecto a otra
#Normalizamos las columnas 8 y 11 (agent y adr) que son las que tienen datos que son muy grandes respecto al resto y lo renombramos como "Normal"
df_normal = df_clean.copy()

scaler = StandardScaler()
df_normal[["agent", "adr"]] = scaler.fit_transform(df_normal[["agent", "adr"]])

df_normal
```
